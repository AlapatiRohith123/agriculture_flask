```python
import pandas as pd
d=pd.read_excel("final catgry.xlsx")
d
```

```python
d=d.drop("Unnamed: 0",axis=1)
```

```python
d=d.drop(['Year_2014 - 2015', 'Year_2015 - 2016','Year_2016 - 2017', 'Year_2017 - 2018', 'Year_2018 - 2019','Year_2019 - 2020'], axis=1)
```

```python
d
```

```python
y=d['Yield (Bales/Hectare)'].values
y
```

```python
x=d.drop('Yield (Bales/Hectare)',axis=1).values
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
model = XGBRegressor(n_estimators=190, learning_rate=0.31)
model.fit(x_train, y_train)
y_pred=model.predict(x_test)
print("r2_score",r2_score(y_test,y_pred))
print("mean_sq_err",mean_squared_error(y_test,y_pred))
```

```python
import pickle as p
p.dump(model,open("model.pkl","wb"))
```
